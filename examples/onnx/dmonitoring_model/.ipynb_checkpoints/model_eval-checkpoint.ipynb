```python
import os
import urllib
import traceback
import time
import sys
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
IMG_PATH = 'ecam.jpeg'
IMG_SIZE = (1440, 960)

REG_SCALE = 0.25
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
class DriverStateV2:
    def __init__(self):
        self.face_orientation = [0, 0, 0]
        self.face_orientation_std = [0, 0, 0]
        self.face_position = [0, 0]
        self.face_position_std = [0, 0]
        self.ready_prob = [0, 0, 0, 0]
        self.not_ready_prob = [0, 0]
        self.face_prob = 0.
        self.left_eye_prob = 0.
        self.right_eye_prob = 0.
        self.left_blink_prob = 0.
        self.right_blink_prob = 0.
        self.sunglasses_prob = 0.
        self.occluded_prob = 0.
```

```python
def parse_driver_data(ds_res, output, out_idx_offset):

    for i in range(3):
        ds_res.face_orientation[i] = output[out_idx_offset + i] * REG_SCALE
        ds_res.face_orientation_std[i] = np.exp(output[out_idx_offset + 6 + i])

    for i in range(2):
        ds_res.face_position[i] = output[out_idx_offset + 3 + i] * REG_SCALE
        ds_res.face_position_std[i] = np.exp(output[out_idx_offset + 9 + i])

    for i in range(4):
        ds_res.ready_prob[i] = sigmoid(output[out_idx_offset + 35 + i])

    for i in range(2):
        ds_res.not_ready_prob[i] = sigmoid(output[out_idx_offset + 39 + i])

    ds_res.face_prob = sigmoid(output[out_idx_offset + 12])
    ds_res.left_eye_prob = sigmoid(output[out_idx_offset + 21])
    ds_res.right_eye_prob = sigmoid(output[out_idx_offset + 30])
    ds_res.left_blink_prob = sigmoid(output[out_idx_offset + 31])
    ds_res.right_blink_prob = sigmoid(output[out_idx_offset + 32])
    ds_res.sunglasses_prob = sigmoid(output[out_idx_offset + 33])
    ds_res.occluded_prob = sigmoid(output[out_idx_offset + 34])
```

```python
img = cv2.imread(IMG_PATH)
img = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
img = img[:, :, 0]
img = cv2.resize(img, IMG_SIZE)
print(img.shape)
plt.imshow(img)
```

```python
img_flipped = cv2.flip(img, 1)
plt.imshow(img_flipped)
```

```python
model_output = np.load("result.npy")[0][0]
model_output_flipped = np.load("result_flipped.npy")[0][0]

# ori
lhd_state = DriverStateV2()
rhd_state = DriverStateV2()
parse_driver_data(lhd_state, model_output, 0)
parse_driver_data(rhd_state, model_output, 41)
poor_vision_prob = sigmoid(model_output[82])
wheel_on_right_prob = sigmoid(model_output[83])

# flipped
lhd_state_flipped = DriverStateV2()
rhd_state_flipped = DriverStateV2()
parse_driver_data(lhd_state_flipped, model_output_flipped, 0)
parse_driver_data(rhd_state_flipped, model_output_flipped, 41)
poor_vision_prob_flipped = sigmoid(model_output_flipped[82])
wheel_on_right_prob_flipped = sigmoid(model_output_flipped[83])
```

```python
# everything about ori
print(lhd_state.face_position, lhd_state.face_prob)
print(rhd_state.face_position, rhd_state.face_prob)
print(lhd_state.left_eye_prob, rhd_state.left_eye_prob)
```

```python
# everything about flipped
print(lhd_state_flipped.face_position, lhd_state_flipped.face_prob)
print(rhd_state_flipped.face_position, rhd_state_flipped.face_prob)
print(lhd_state_flipped.left_eye_prob, rhd_state_flipped.left_eye_prob)
```

```python
# to eval ori and flipped
print(wheel_on_right_prob, wheel_on_right_prob_flipped)
```

```python
# using ori data
face_dx = lhd_state_flipped.face_position[0]
face_dy = lhd_state_flipped.face_position[1]
alpha = 1.0  # Example value for alpha
box_size = 220  # Example value for box_size

image_width = 1440.
image_height = 960.

# Calculate center_x and center_y
#fcenter_x = int(image_width / 2 + face_dx * image_width / 2)
fcenter_x = int(1440.0 - 1714.0 * face_dx)
#fcenter_y = int(image_height / 2 + face_dy * image_height / 2)
fcenter_y = int(-135.0 + (504.0 + abs(face_dx)*112.0) + (1205.0 - abs(face_dx)*724.0) * face_dy)

# Create a white pen with the specified alpha and thickness
color = (255, 255, 255)
thickness = 20
drawn_img = cv2.rectangle(img, (fcenter_x - box_size // 2, fcenter_y - box_size // 2), (fcenter_x + box_size // 2, fcenter_y + box_size // 2), color, thickness)

# Display the resulting image
plt.imshow(drawn_img)
```
